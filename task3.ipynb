```python
from typing import TypedDict, Annotated, List, Dict, Any
import operator
from langgraph.graph import StateGraph, END
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
from langchain_google_genai import ChatGoogleGenerativeAI
from google.colab import userdata





llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    google_api_key=userdata.get("napi")
)
```

```python
from typing import TypedDict

class SoftwareAgencyState(TypedDict):
    user_requirement: str

    project_type: str

    requirement_analysis: str
    architecture_plan: str

    frontend_code: str
    backend_plan: str
    backend_code: str

    database_plan: str
    database_schema: str

    qa_plan: str

    review_report: str

    final_report: str
```

```python
def requirement_analyzer(state):

    prompt = f"""
    Analyze the following software requirement.

    Requirement:
    {state['user_requirement']}

    Extract:

    1. Project Name
    2. Main Features
    3. User Roles
    4. Complexity Level
    5. Functional Requirements

    Return detailed analysis.
    """

    response = llm.invoke(prompt)

    return {
        "requirement_analysis": response.content
    }
```

```python
def project_classifier(state):

    req = state["user_requirement"].lower()

    ai_keywords = [
        "ai",
        "machine learning",
        "llm",
        "rag",
        "chatbot",
        "interview assistant"
    ]

    web_keywords = [
        "website",
        "portfolio",
        "ecommerce",
        "shopping",
        "landing page"
    ]

    enterprise_keywords = [
        "erp",
        "hospital",
        "bank",
        "college",
        "university",
        "healthcare"
    ]

    if any(word in req for word in ai_keywords):
        project_type = "AI"

    elif any(word in req for word in web_keywords):
        project_type = "Web"

    elif any(word in req for word in enterprise_keywords):
        project_type = "Enterprise"

    else:
        project_type = "Web"

    return {
        "project_type": project_type
    }
```

```python
def route_project(state):
    return state["project_type"]
```

```python
def ai_architect(state):

    prompt = f"""
    You are an AI Solution Architect.

    Requirement:
    {state['user_requirement']}

    Generate:

    1. System Architecture
    2. AI Components
    3. LLM Design
    4. Vector Database
    5. Tech Stack
    6. Folder Structure
    7. Sample Python Code
    """

    response = llm.invoke(prompt)

    return {
        "architecture_plan": response.content,
        "frontend_code": response.content
    }
```

```python
def frontend_architect(state):

    prompt = f"""
    You are a Senior Frontend Architect.

    Requirement:
    {state['user_requirement']}

    Generate:

    1. Website Architecture
    2. Pages
    3. Components
    4. React Folder Structure
    5. Tailwind Design
    6. React Sample Code
    """

    response = llm.invoke(prompt)

    return {
        "architecture_plan": response.content,
        "frontend_code": response.content
    }
```

```python
def enterprise_architect(state):
    print("Enterprise architect function called!")
    prompt = f"""
    You are an Enterprise Software Architect.

    Requirement:
    {state['user_requirement']}

    Generate:

    1. Enterprise Architecture
    2. Modules
    3. User Roles
    4. Microservices Design
    5. Folder Structure
    6. UI Layout
    7. Sample Module Code
    """

    response = llm.invoke(prompt)

    return {
        "architecture_plan": response.content,
        "frontend_code": response.content
    }
```

```python
def backend_agent(state):

    prompt = f"""
    You are a Backend Architect.

    Requirement:
    {state['user_requirement']}

    Architecture:
    {state['architecture_plan']}

    Generate:

    1. REST APIs
    2. Authentication Strategy
    3. Express Folder Structure
    4. Backend Workflow
    5. Node.js Sample Code
    """

    response = llm.invoke(prompt)

    return {
        "backend_plan": response.content,
        "backend_code": response.content
    }
```

```python
def database_agent(state):

    prompt = f"""
    You are a Database Architect.

    Requirement:
    {state['user_requirement']}

    Generate:

    1. Tables
    2. Relationships
    3. Primary Keys
    4. Foreign Keys
    5. Indexes
    6. SQL Schema

    Return complete SQL code.
    """

    response = llm.invoke(prompt)

    return {
        "database_plan": response.content,
        "database_schema": response.content
    }
```

```python
def qa_agent(state):

    prompt = f"""
    You are a QA Engineer.

    Requirement:
    {state['user_requirement']}

    Generate:

    1. Functional Test Cases
    2. Integration Test Cases
    3. Security Test Cases
    4. Edge Cases
    5. Performance Tests
    """

    response = llm.invoke(prompt)

    return {
        "qa_plan": response.content
    }
```

```python
def reviewer_agent(state):

    prompt = f"""
    Review the following outputs.

    Architecture:
    {state['architecture_plan']}

    Backend:
    {state['backend_plan']}

    Database:
    {state['database_plan']}

    QA:
    {state['qa_plan']}

    Return:

    APPROVED

    or

    NEEDS_REVISION

    with explanation.
    """

    response = llm.invoke(prompt)

    return {
        "review_report": response.content
    }
```

```python
def final_report_agent(state):

    report = f"""
# AI SOFTWARE DEVELOPMENT AGENCY REPORT

=======================================

PROJECT TYPE
{state['project_type']}

=======================================

REQUIREMENT ANALYSIS

{state['requirement_analysis']}

=======================================

ARCHITECTURE

{state['architecture_plan']}

=======================================

FRONTEND

{state['frontend_code']}

=======================================

BACKEND

{state['backend_plan']}

=======================================

DATABASE

{state['database_schema']}

=======================================

QA

{state['qa_plan']}

=======================================

REVIEW

{state['review_report']}
"""

    return {
        "final_report": report
    }
```

```python
from langgraph.graph import StateGraph, END
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
from langchain_google_genai import ChatGoogleGenerativeAI
from google.colab import userdata

workflow = StateGraph(SoftwareAgencyState)

workflow.add_node(
    "requirement_analyzer",
    requirement_analyzer
)

workflow.add_node(
    "project_classifier",
    project_classifier
)

workflow.add_node(
    "ai_architect",
    ai_architect
)

workflow.add_node(
    "frontend_architect",
    frontend_architect
)

workflow.add_node(
    "enterprise_architect",
    enterprise_architect
)

workflow.add_node(
    "backend_agent",
    backend_agent
)

workflow.add_node(
    "database_agent",
    database_agent
)

workflow.add_node(
    "qa_agent",
    qa_agent
)

workflow.add_node(
    "reviewer_agent",
    reviewer_agent
)

workflow.add_node(
    "final_report_agent",
    final_report_agent
)

workflow.set_entry_point(
    "requirement_analyzer"
)

workflow.add_edge(
    "requirement_analyzer",
    "project_classifier"
)

workflow.add_conditional_edges(
    "project_classifier",
    route_project,
    {
        "AI": "ai_architect",
        "Web": "frontend_architect",
        "Enterprise": "enterprise_architect"
    }
)

workflow.add_edge(
    "ai_architect",
    "backend_agent"
)

workflow.add_edge(
    "frontend_architect",
    "backend_agent"
)

workflow.add_edge(
    "enterprise_architect",
    "backend_agent"
)

workflow.add_edge(
    "backend_agent",
    "database_agent"
)

workflow.add_edge(
    "database_agent",
    "qa_agent"
)

workflow.add_edge(
    "qa_agent",
    "reviewer_agent"
)

workflow.add_edge(
    "reviewer_agent",
    "final_report_agent"
)

workflow.add_edge(
    "final_report_agent",
    END
)

graph = workflow.compile()
```

```python
result = graph.invoke(
    {
        "user_requirement":
        "Build a College ERP System"
    }
)

print(result["final_report"])
```
